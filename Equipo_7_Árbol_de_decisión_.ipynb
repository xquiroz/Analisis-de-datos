### Arbol de decisiones


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
columnas= ["age", "workclass", "fnlwgt", "education", "education_num", "marital_status", "occupation", "relationship", "race", "sex", "capital_gain", "capital_loss", "hours_per_week", "native_country", "income"]
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', names=columnas)
data.head()#visualizacion de las primeras filas de nuestro dataset
```

```python
data.dropna(inplace=True) #eliminamos datos faltantes
```

```python
#se busca predecir el income dado el sexo y los años estudiado de las personas,
# entonces solo nos quedamos con esos datos
data= data[["sex", "education_num", "income"]]
data.head()#visualizamos lod datos
```

```python
#ahora convertimos las variables sex e income en binarias
data["sex"] = (data["sex"] == " Female").astype(int)#si es mujer se marca con 1
data["income"] = (data["income"] == ' >50K').astype(int)#mayor a 50k se marca con 1
data
```

```python
data.info()
```

```python
#dividimos los datos en entrenamiento y prueba
X = data.drop('income', axis=1)  #variables independientes
y = data['income'] #variable dependiente
```

```python
from sklearn.model_selection import train_test_split
#75% para entrenamiento, 25% test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Crear un modelo de Árbol de Decisiones
arbol_decision = DecisionTreeClassifier()

# Entrenar el modelo con los datos de entrenamiento
arbol_decision.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = arbol_decision.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Mostrar la matriz de confusión y el informe de clasificación
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```

```python
from sklearn.tree import export_graphviz
import graphviz

# Crear una representación gráfica del árbol
dot_data = export_graphviz(arbol_decision, out_file=None,
                            feature_names=["sex", "education_num"],
                            class_names=["0", "1"],
                            filled=True, rounded=True, special_characters=True)

# Visualizar el árbol
graph = graphviz.Source(dot_data)
graph.render("arbol_de_decision")  # Guardar el gráfico como un archivo, por ejemplo, "arbol_de_decision.pdf"


```
